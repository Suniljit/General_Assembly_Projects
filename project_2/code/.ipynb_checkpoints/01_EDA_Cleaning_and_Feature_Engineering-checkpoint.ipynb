# Project 2: Regression Challenge (HDB Prices)

## 01: EDA, Cleaning & Feature Engineering

### Problem Statement

We are representing a property research company that aims to predict accurate HDB resale prices as well as understand the factors that affect the resale price. This would serve as a valuable reference for real estate agents and home buyers to enable them to make an informed decision.  

For our purposes, we will develop a linear regression model as they are highly interpretable. This interpretability would be important as we need to access the factors that affect the resale price. Futhermore, we will try multiple models, including vanilla linear regression, and linear regression models with regularization such as ridge, lasso and elastic net.    

We will evaluate the effectiveness of the model by looking at the root mean squared error.  

Based on the results observed, we will choose the best model to use as our production model and apply it to unseen data.

### Prelimary columns to drop using domain knowledge

The following columns are dropped using common knowledge/logic as well as domain knowledge

| Columns to drop | Rationale |
|-----------------|-----------|
|id |Does not provide any meaningful information other than as an identifier|
|block, street_name, address, postal, Latitude, Longitude |We will be using the town column to provide a more generalize locale rather than specific locations|
|Tranc_YearMonth |This is dropped as there are already two columns for Tranc_Year and Tranc_Month|
|full_flat_type | This information is already provided in two seperate columsn (flat_type and flat_model)|
|planning_area |We will use the town column for the generale locale as these are resale flats, and the town column would provide a more up to date information to buyers/sellers compared to the planning area|
|mrt_name, mrt_latitude, mrt_longitude|The name of the MRT station and location bears no significance. What matters is how far away the station is|
|bus_stop_name, bus_stop_latitude, bus_stop_longitude| The name of the bus stop and location bears no significance. What matters is how far away the bus stop is|
|pri_sch_latitude, pri_sch_longitude, sec_sch_latitude, sec_sch_longitude | The exact location of the schools bears no significance|
|pri_sch_name, vacancy, pri_sch_affiliation, sec_sch_name, cutoff_point, affiliation|As we are looking for flats to hold until the price appreciates, we're not too concerned about the schools nearby as school relocation and mergers are becoming quite common. Hence, the current schools in the area are not an important metric. However, we will keep pri_sch_nearest_distance and sec_sch_nearest_dist to ensure that there is a school nearby for families. Similarly, the school's vacancy, affiliations and cutoff points are inconsequential as we're looking beyong a one-year timeframe|  
|storey_range, lower, upper, mid_storey|We will just use mid, which is the same mid_storey value (due to the nature of the dataset), and is the middle of storey_range. Upper and lower values are just +1 and -1 of the mid_storey value. Hence, these other columns are unneccesary|
|hdb_age|As the y-value we're looking to train and predict is the resale price, the year of transaction bears more significant compared to the present age of the flat|
|year_completed|This column will be dropped as in Singapore, the lease remaining duration is more important. A feature for that will be engineered later on. Furthermore, purchasing an old flat in the hopes of the government repurchasing it (Enbloc) would be equivalent to gambling, which is not helpful for the analysis here|
|Mall_Within_500m, Mall_Within_1k, Mall_Within_2km, Hawker_Within_500m, Hawker_Within_1km, Hawker_Within_2km|These columns have a significant amount of missing values, which prevents us from dropping the rows with missing values. Imputing the values would result in most rows having the same value, which provides little benefit|
|floor_area_sqft|Exact same information as floor_area_sqm, just in a different metric. As Singapore uses meters as a metric, we will drop the one with square foot|

```python
# Drop the columns as specified above
hdb.drop(columns=['id', 'block', 'street_name', 'address', 'postal', 
                  'Latitude', 'Longitude', 'Tranc_YearMonth', 'full_flat_type',   
                 'planning_area', 'mrt_name', 'mrt_latitude', 'mrt_longitude',
                 'bus_stop_name', 'bus_stop_latitude', 'bus_stop_longitude', 
                 'pri_sch_latitude', 'pri_sch_longitude', 'sec_sch_latitude',
                 'sec_sch_longitude', 'pri_sch_name', 'vacancy', 
                  'pri_sch_affiliation', 'sec_sch_name', 'cutoff_point', 
                  'affiliation','storey_range', 'lower', 'upper', 
                  'mid_storey', 'hdb_age', 'year_completed', 'Mall_Within_500m', 
                 'Mall_Within_1km', 'Mall_Within_2km', 'Hawker_Within_500m',
                 'Hawker_Within_1km', 'Hawker_Within_2km', 'floor_area_sqft'], 
         inplace=True)
```

#### Columns with binary data

Certain columns have binary data (e.g. 0/1 or Y/N). If majority of the entries are skewed to one value (e.g. 99% are Y, 1% are N), the column would not be able to provide much insight and should be droped. In the next step, we will check for this.

```python
# Look at columns binary data (0/1 or Y/N) to see if they would be able to contribute as features

cols_binary = hdb[['residential', 'commercial', 'market_hawker', 
                  'multistorey_carpark', 'precinct_pavilion', 
                 'bus_interchange', 'mrt_interchange']]

for col in cols_binary:
    print(f"The distribtuion of {col} is: ")
    print(hdb[col].value_counts(normalize=True))
    print()
```

From the distributions above, we can see that the columns residential, market_hawker, multistorey_carpark, precinct_pavilion will provide no benefit as a feature as over 99% of their values are the same. Hence, they will be dropped.

#### Rental flat data

In this section, we'll look into the rental flat data to see if they will be able to provide any value to the model.

There are nearly 9000 unique values ranging from 0m to 3497m for the column Mall_Nearest_Distance. Hence, let's impute with median to fill the missing values.

```python
# Impute the missing values with median
median_Mall_Nearest_Distance = np.median(hdb['Mall_Nearest_Distance'].dropna())
hdb['Mall_Nearest_Distance'] = hdb['Mall_Nearest_Distance'].fillna(median_Mall_Nearest_Distance)
```

#### Feature Engineering

```python
# Create a column for lease_remaining_years and drop lease_commence_date
hdb['lease_remaining_years'] = 99 - (2021 - hdb['lease_commence_date'])
hdb.drop(columns='lease_commence_date', inplace=True)
```

```python
# Check if the column is created correctly
hdb['lease_remaining_years'].head()
```

**Features: hawker_food_stalls & hawker_market_stalls**  

With regards to these two features, the number of stalls itself should not matter, as typically, a few stalls (approx. 5) would be more than sufficient. Hence, looking at the number of stalls present would not be valuable. Rather, the important consideration when someone is looking to buy a home is if a hawker food stall and market is available. As such, we will engineer two new columns with binary values to state if food or market stalls are present. 

```python
# Convert all values greater than 0 to 1
hdb['hawker_food_stalls'] = [1 if x > 0 else 0 for x in 
                             hdb['hawker_food_stalls']]
```

```python
# Check the distribution of the values
hdb['hawker_food_stalls'].value_counts()
```

From the distribution, we can see that almost all hawker centres have food stalls. Hence, keeping this feature would not provide any value to the model. We will drop this feature. 

```python
# Drop hawker_food_stalls
hdb.drop(columns='hawker_food_stalls', inplace=True)
```

```python
# Convert all values greater than 0 to 1
hdb['hawker_market_stalls'] = [1 if x > 0 else 0 for x in 
                               hdb['hawker_market_stalls']]
```

```python
# Check the distribution of the values
hdb['hawker_market_stalls'].value_counts()
```

From the distribution, we can see that this feature will be able to provide valuable information about whether market stalls are available to people in the estate.

### Correlation of features (check for multi-collinearity)

In this section, we'll plot the correlation heatmap to check for multi-collinearity.

```python
# Establish size of figure.
plt.figure(figsize = (25,16))

# Title
plt.title("Correlation of HDB data")

# Drop the categorical varialbes and get the correlation of the numerical variables.
corr_hdb = hdb.drop(columns = ['town', 'flat_type', 'flat_model', 
                               'commercial', 'bus_interchange', 
                              'mrt_interchange']).corr()

# Set up mask to be "True" in the upper triangle.
mask = np.zeros_like(corr_hdb)
mask[np.triu_indices_from(mask)] = True

# Plot our correlation heatmap, while masking the upper triangle to be white.
with sns.axes_style("white"):
    sns.heatmap(corr_hdb, mask = mask, square = True, annot = True, vmin = -1, 
                vmax = 1, linewidths = .5, cmap='vlag')
```

#### Analysis of High Correlations (excluding resale price)

We shall consider any correlation > 0.5 as high and look into it further for multicolinearity 

Summary of high correlations (excluding resale price)
|Col 1| Col 2| Corr| Analysis|

|mid| max_floor_lvl| 0.56| Both features move the same with respect to the resale price, hence we will drop max_floor_lvl|
|floor_area_sqm| exec_sold| 0.51| No significant relationship found. We will keep both features|
|floor_area_sqm| 3room_sold| -0.59| No significant relationship found. We will keep both features|
|total_dwelling_units| 3rooms_sold| 0.68| No significant relationship found. We will keep both features|
|3room_sold| lease_remaining_years| -0.5|  No significant relationship found. We will keep both features|

```python
# Plot a regression line between resale price and max_floor_lvl/mid
fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(16, 9))
sns.regplot(hdb, x='max_floor_lvl', y='resale_price', scatter=False, ax=ax1)
sns.regplot(hdb, x='mid', y='resale_price', scatter=False, ax=ax2);
```

Correlation can be clearly observed between the mid_storey sold vs max_floor_lvl. This is understandable as the higher the max_floor_lvl is, a higher storey flat can be sold, which allows the mid_storey value to increase as well. From the linear regression plots above, we can see that both features move exactly the same way when compared to the resale price. Hence, one of these features can be dropped. As mid_storey can directly relate to the resale price compared to max_floor_lvl of the block. we will keep mid_storey and drop max_floor_lvl
